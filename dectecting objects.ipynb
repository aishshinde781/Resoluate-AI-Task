# Import Required Libraries

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

we will read the image and then convert this image into grayscale image

```python
image = cv.imread("freezer_image (1).jpg")
```

```python
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

For counting, we have to detect the edges but before detecting the edges we have to make the image blur to avoid the noise.

```python
blur = cv.GaussianBlur(gray,(11,11),0)
plt.imshow(blur, cmap='gray')
```

Now we will detect the edges using canny algorithm.

```python
canny = cv.Canny(blur,30,150,3)
plt.imshow(canny,cmap='gray')
```

We can see that edges are not connected. We need to connect the edges,have to make more thicker and visible.

```python
dilated = cv.dilate(canny,(1,1),iterations=0)
plt.imshow(dilated,cmap='gray')
```

Now we have to calculate the contour in the image and convert the image into RGB from BGR and the draw the contours.

```python
(cnt, hierarchy) = cv.findContours(
    dilated.copy(), cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)
rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)
cv.drawContours(rgb, cnt, -1, (0, 255, 0), 2)
  
plt.imshow(rgb)
```
